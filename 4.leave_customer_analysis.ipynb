```python
import pandas as pd
import matplotlib.pyplot as plt
```

Clustering 전처리 결과와 기존 데이터셋을 결합하여 구매하지 않은 이탈고객에 대한 분석 진행

```python
df = pd.read_csv('cluster_5.csv')
```

```python
df.head()
```

```python
df_g1 = df[df['labels5'] == 1]
```

```python
df_g1
```

```python
df_g1_pur = df_g1[df_g1['event_type_purchase'] > 0]
```

```python
df_g1_pur
```

```python
dfo = df[['user_id','labels5']]
```

```python
elect = pd.read_csv('electonic/events.csv')
```

```python
elect = elect[elect['category_code'].str.contains('electronics', na=False)]
elect = elect.dropna(subset=['brand', 'user_session']) 
```

```python
elect.head()
```

```python
df_group = pd.merge(elect, dfo, on='user_id', how = 'inner')
```

구매건에 대한 분석

```python
df_pur = df_group[df_group['event_type'] == 'purchase']
```

```python
df_pur.groupby('labels5')['price'].agg('mean')
```

구매한 고객의 구매 금액(합계, 평균, 구매 수)

```python
df_pur.groupby('labels5')['price'].agg(['sum','count','mean'])
```

구매 유저 수

그룹별 구매 건당 평균 금액 시각화

```python
pur_price = pd.DataFrame(df_pur.groupby('labels5')['price'].agg('mean'), columns=['price'])
```

```python
plt.rc('font', size=15)
```

```python
plt.barh(pur_price.index, pur_price.price, color = ['gray', 'yellow', 'gray', 'gray', 'gray'])
for idx, value in enumerate(pur_price.price):
    if idx == 1:
        plt.text(value-250, idx, str(round(value, 2)))
    else:
        plt.text(value+0.5, idx, str(round(value, 2)))
```

그룹 3 view, purchase 건수 별 유저 분석

```python
df_group3 = df_group[df_group['labels5'] == 3]
```

```python
df_group3[df_group3['event_type'] == 'purchase'].groupby(['category_code'])['price'].agg('mean')
```

```python
df3_view_user = pd.DataFrame(df_group3[df_group3['event_type'] == 'view'].groupby(['category_code'])['user_id'].agg('nunique'),columns=['user_id'])
```

```python
df3_view_user.sort_values(by = 'user_id', ascending=True, inplace = True)
```

```python
df3_view_user['percentile'] = round(df3_view_user['user_id'] / sum(df3_view_user['user_id']) * 100, 2)
```

```python
df3_view_user
```

```python
df3_purchase_price = pd.DataFrame(df_group3[df_group3['event_type'] == 'purchase'].groupby(['category_code'])['price'].agg('sum'), columns=['price'])
```

```python
df3_purchase_price.sort_values(by = 'price', ascending=True, inplace = True)
```

```python
df3_purchase_price['percentile'] = round(df3_purchase_price['price'] / sum(df3_purchase_price['price']) * 100, 2)
```

```python
plt.rc('font', size=10)
```

```python
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
```

```python
plt.barh(df3_view_user.index, df3_view_user.user_id)
for idx, value in enumerate(df3_view_user.user_id):
    plt.text(value+0.5, idx, str(round(value, 2)))
plt.xlabel('고객수(명)')
```

```python
plt.barh(df3_purchase_price.index, df3_purchase_price.price)
for idx, value in enumerate(df3_purchase_price.price):
    if idx == 11:
        plt.text(value-10000, idx, str(round(value, 2))+ f'({df3_purchase_price.percentile[idx]}%)', color = 'white')
    else:
        plt.text(value+0.5, idx, str(round(value, 2))+ f'({df3_purchase_price.percentile[idx]}%)')
plt.xlabel('매출')
```
